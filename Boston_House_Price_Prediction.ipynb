```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression , Ridge
from sklearn.metrics import r2_score,mean_squared_error
```

```python
df=pd.read_csv('Boston Dataset.csv')
```

```python
target_feature='medv'
y=df[[target_feature]]
X=df.drop(target_feature,axis=1)
```

```python
y
```

```python
X.head()
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.18,shuffle=True,random_state=45)
```

```python
rf=RandomForestRegressor(n_estimators=90,random_state=45) #First we are testing the Random forest regressor
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred_rf=rf.predict(X_test)
```

```python
print('Mean squared error = ',round(mean_squared_error(y_test,y_pred_rf),3))
print('R square score = ',round(r2_score(y_test,y_pred_rf),3))
```

As we can see above given R square score is 0.912 which is great for our model as compared to other normal model which we will see below

```python
lr=LinearRegression()#Now we are doing the same thing with Linear regression to check whether the random forest regressor gave good ouputs compared to it or not
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_lr=lr.predict(X_test)
```

```python
print('Mean squared error = ',round(mean_squared_error(y_test,y_pred_lr),3))
print('R square score = ',round(r2_score(y_test,y_pred_lr),3))
```

As we can see here that the r square score is just 0.736 and the random forest had much better accuracy then this model , hence we have successfully made a house price prediction model for boston.
